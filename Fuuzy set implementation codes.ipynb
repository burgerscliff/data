```python
import numpy as np
import matplotlib.pyplot as plt

# Set up the visual style for the plots
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
def trimf(x, a, b, c):
    """Triangular Membership Function"""
    # Using np.maximum and np.minimum to cleanly handle the piece-wise logic
    # Adding a tiny value (1e-15) to avoid division by zero errors if a=b or b=c
    left_slope = (x - a) / (b - a + 1e-15)
    right_slope = (c - x) / (c - b + 1e-15)
    return np.maximum(0, np.minimum(left_slope, right_slope))

def trapmf(x, a, b, c, d):
    """Trapezoidal Membership Function"""
    left_slope = (x - a) / (b - a + 1e-15)
    right_slope = (d - x) / (d - c + 1e-15)
    return np.maximum(0, np.minimum(np.minimum(left_slope, 1.0), right_slope))

def gaussmf(x, mean, sigma):
    """Gaussian Membership Function"""
    return np.exp(-((x - mean) ** 2) / (2 * sigma ** 2))

def gbellmf(x, a, b, c):
    """Generalized Bell Membership Function"""
    return 1.0 / (1.0 + np.abs((x - c) / a) ** (2 * b))

def sigmf(x, slope, center):
    """Sigmoidal Membership Function"""
    return 1.0 / (1.0 + np.exp(-slope * (x - center)))
```

```python
# Create the Universe of Discourse (x-axis values from 0 to 10)
x = np.linspace(0, 10, 500)

print("--- Fuzzy Membership Function ---")
print("1. Triangular")
print("2. Trapezoidal")
print("3. Gaussian")
print("4. Generalized Bell")
print("5. Sigmoidal")

choice = input("Select a function to plot (Enter 1-5): ")

plt.figure(figsize=(8, 5))
valid_choice = True

try:
    if choice == '1':
        print("\nTriangular Function needs 3 points: lower bound (a), peak (b), upper bound (c).")
        a = float(input("Enter a (e.g., 2): "))
        b = float(input("Enter b (e.g., 5): "))
        c = float(input("Enter c (e.g., 8): "))
        y = trimf(x, a, b, c)
        plt.plot(x, y, 'b-', linewidth=2)
        plt.title(f'Triangular Membership Function\n(a={a}, b={b}, c={c})')

    elif choice == '2':
        print("\nTrapezoidal Function needs 4 points: bottom-left (a), top-left (b), top-right (c), bottom-right (d).")
        a = float(input("Enter a (e.g., 1): "))
        b = float(input("Enter b (e.g., 3): "))
        c = float(input("Enter c (e.g., 7): "))
        d = float(input("Enter d (e.g., 9): "))
        y = trapmf(x, a, b, c, d)
        plt.plot(x, y, 'g-', linewidth=2)
        plt.title(f'Trapezoidal Membership Function\n(a={a}, b={b}, c={c}, d={d})')

    elif choice == '3':
        print("\nGaussian Function needs 2 parameters: center/mean and standard deviation/width (sigma).")
        mean = float(input("Enter center/mean (e.g., 5): "))
        sigma = float(input("Enter width/sigma (e.g., 1.5): "))
        y = gaussmf(x, mean, sigma)
        plt.plot(x, y, 'r-', linewidth=2)
        plt.title(f'Gaussian Membership Function\n(mean={mean}, sigma={sigma})')

    elif choice == '4':
        print("\nGeneralized Bell Function needs 3 parameters: width (a), shape (b), center (c).")
        a = float(input("Enter width 'a' (e.g., 2): "))
        b = float(input("Enter shape 'b' (e.g., 4): "))
        c = float(input("Enter center 'c' (e.g., 5): "))
        y = gbellmf(x, a, b, c)
        plt.plot(x, y, 'm-', linewidth=2)
        plt.title(f'Generalized Bell Membership Function\n(a={a}, b={b}, c={c})')

    elif choice == '5':
        print("\nSigmoidal Function needs 2 parameters: slope (positive for S, negative for Z) and crossover center.")
        slope = float(input("Enter slope (e.g., 2): "))
        center = float(input("Enter crossover center (e.g., 5): "))
        y = sigmf(x, slope, center)
        plt.plot(x, y, 'c-', linewidth=2)
        plt.title(f'Sigmoidal Membership Function\n(slope={slope}, center={center})')

    else:
        print("\nInvalid selection. Please run the cell again and enter a number from 1 to 5.")
        valid_choice = False

    if valid_choice:
        plt.ylim([-0.05, 1.05])
        plt.ylabel('Membership Degree \u03bc(x)')
        plt.xlabel('Universe of Discourse (x)')
        plt.show()

except ValueError:
    print("\nError: Please enter valid numbers only.")
```
